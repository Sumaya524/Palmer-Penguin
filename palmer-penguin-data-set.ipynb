Both Datasets contain data for 344 penguins. There are 3 different species of penguins in this dataset , collected from 3 islands in the Palmer Archipelago, Antartica

Penguins are fun to summarize

```r
penguins %>%
count(species)
```

The penguin dataset contains size measurments for three penguin species (Adelie , Chinstrap, Gentoo)

The column include the information such as body mass, flipper length and bill length

# Visualize

Plotting the relationship between two variables flipper length and body mass

```r
ggplot(data=penguins) +geom_point(mapping=aes(x=flipper_length_mm, y=body_mass_g))
```

Here plotting the relationship between bill length and bill depth

```r
ggplot(data=penguins) +geom_point(mapping=aes(x=bill_length_mm, y=bill_depth_mm))
```

We want to show the relationship between the trendline and data point more clearly

```r
ggplot(data=penguins) + geom_smooth(mapping=aes(x=flipper_length_mm, y=body_mass_g)) +
geom_point(mapping=aes(x=flipper_length_mm, y=body_mass_g))
```

We want  to plot a separate line for each species of penguin

```r
ggplot(data=penguins) + geom_smooth(mapping=aes(x=flipper_length_mm, y=body_mass_g, linetype= species))
```

geom_jitter creates ascatter plot, jittering help us deal with over- plotting

```r
ggplot(data=penguins) + geom_jitter(mapping=aes(x=flipper_length_mm, y=body_mass_g))
```

```r
ggplot(data=penguins) + geom_point(mapping=aes(x=flipper_length_mm, y=body_mass_g, color=species)) +
labs(title="Palmer Penguins : Body Mass Vs. Flipper Length", caption="Data collected by Dr. Kristen Gorman") + 
annotate("text", x=220, y= 3500,label =" Gentoos are the largest", color="purple", fontface="bold" , size =4.5, angle =25)
```

```r
ggplot(data=penguins) + geom_point(mapping=aes(x=flipper_length_mm, y=body_mass_g, shape=species, color=species, size=species))
```

```r
ggplot(data=penguins) + geom_point(mapping=aes(x=flipper_length_mm, y=body_mass_g, color=species)) +
facet_wrap(~ species) + labs(title="Palmer Penguins : Body Mass Vs. Flipper Length", subtitle ="Flipper Length and body mass for Three penguins")
```

```r
ggplot(data=penguins) + geom_point(mapping=aes(x=flipper_length_mm, y=body_mass_g, color=species)) +
facet_grid(sex~species) + labs(title="Palmer Penguins : Body Mass Vs. Flipper Length", subtitle ="Flipper Length and body mass for Three penguins")
```
